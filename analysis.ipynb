### Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read in data

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
# Order the Stay_In_Current_City_Years variable
df['Stay_In_Current_City_Years'] = pd.Categorical(df['Stay_In_Current_City_Years'], ['0','1','2','3', '4+'])
```

```python
# Order the age variable
df['Age'] = pd.Categorical(df['Age'], ['0-17','18-25','26-35','36-45', '46-50', '51-55', '55+'])
```

# Task 1 Diagrams 

## 1.1. Draw an appropriate diagram to represent the percentage distribution of all the product categories.

```python
# Combine all three category columns into a single Series
all_categories = pd.concat([df['Product_Category_1'], df['Product_Category_2'], df['Product_Category_3']])

# Calculate the percentage distribution of each category
category_counts = all_categories.value_counts(normalize=True) * 100

# Create a barplot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index, y=category_counts.values, palette="viridis")

# Customize the plot
plt.title("Percentage Distribution of Product Categories")
plt.xlabel("Product Categories")
plt.ylabel("Percentage (%)")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha="right")

# Show the plot
plt.tight_layout()
plt.show()
```
